## Câu 1 (2.5 đ)

Tạo mảng một chiều NumPy `a` gồm các số nguyên từ 1 đến 20. Thực hiện các thao tác sau trên `a`:

1. Cho biết 5 số cuối cùng của mảng.

1. Trích ra các số chẵn trong mảng.

1. Trích ra các số lẻ lớn hơn 10 trong mảng.

1. Tạo mảng `b` gồm phần đầu là các số chẵn và phần sau là các số lẻ trong `a`.

1. Tảo mảng `c` 2 chiều gồm dòng đầu là các số chẵn và dòng cuối là các số lẻ trong `a`.

1. Tảo mảng `d` 2 chiều gồm cột đầu là các số chẵn và cột cuối là các số lẻ trong `a`.

1. Thay tất cả các số chia hết cho 5 trong `a` thành số 0.

1. Xóa tất cả các số 0 trong `a`.

1. Đảo ngược thứ tự các phần tử trong `a`.

1. Sắp theo thứ tự tăng dần các số trong `a`.

```python
# Create an array NumPy a with integers from 1 to 20
a = np.arange(1, 21)

# 1. Last 5 elements of a
print(f"1. 5 phan tu cuoi cung: {a[-5:]}")

# 2. Print all even numbers from a
print(f"2. So chan trong mang: {a[a % 2 == 0]}")

# 3. Print all odd numbers and greater than 10 from a
print(f"3. So le trong mang va lon hon 10: {a[(a % 2 != 0) & (a > 10)]}")

# 4. Create a NumPy array b of the first even numbers and the last odd numbers of a
b = np.concatenate((a[a % 2 == 0], a[a % 2 != 0]))
print(f"4. Mang b: {b}")

# 5. Create a 2D NumPy array c with first row is even numbers of a and second row is odd numbers of a
c = np.array([a[a % 2 == 0], a[a % 2 != 0]])
print(f"5. Mang c: {c}")

# 6. Create a 2D NumPy array d with first column is even numbers of a and second column is odd numbers of a
d= c.transpose()
print(f"6. Mang d: {d}")

# 7. Replace all number devide by 5 of a by 0
a[a % 5 == 0] = 0
print(f"7. Thay tat ca so chia het cho 5 bang so 0. KQ a: {a}")

# 8. Remove all 0 of a
a = a[a != 0]
print(f"8. Xoa tat ca so 0. KQ a: {a}")

# 9. Reverse a
a = a[::-1]
print(f"9. Dao nguoc mang a: {a}")

# 10. Sort a
a.sort()
print(f"10. Sap xep mang a: {a}")
```

## Câu 2 (1.5 đ)

Tạo `mat` là một ma trận kích thước 3x4 gồm các số nguyên với phần tử ở dòng $i$, cột $j$ là $i \times j$ (chỉ số dòng, cột tính từ 1).

1. Xuất ra dòng đầu của `mat`.

1. Xuất ra phần tử ở dòng cuối, cột đầu của `mat`.

1. Xuất ra nửa phải của `mat`.

1. Đặt tất cả các phần tử dòng đầu của `mat` là 0.

1. Tách `mat` theo chiều dọc thành 2 ma trận con 2 cột.

1. Tách `mat` theo chiều ngang thành 2 ma trận con lần lượt gồm 1 dòng và 2 dòng.

```python
# Create mat is a 3x4 matrix. Each value represents the row * column of the element, count from 1 (E.g [0,0] is 1*1=1.)
# [1, 2, 3, 4]
# [2, 4, 6, 8]
# [3, 6, 9, 12]
rows = 3
cols = 4
# ref: https://stackoverflow.com/questions/477486/how-to-use-a-decimal-range-step-value
# ref: https://stackoverflow.com/questions/6667201/how-to-define-two-dimensional-array-in-python
# reshape: https://stackoverflow.com/questions/18691084/what-does-1-mean-in-numpy-reshape
i = np.arange(1, rows + 1).reshape(rows, 1)
j = np.arange(1, cols + 1)
mat = i * j

print(f"mat: {mat}")

# 1. Print the first row of mat
print(f"1. Dong dau tien cua mat: {mat[0]}")
# 2. Print the element at the end of the last row and first column of mat
print(f"2. Phan tu cuoi cung cua dong cuoi cung va cot dau tien cua mat: {mat[-1, 0]}")
# 3. Print the right half of mat
print(f"3. Nua phai cua mat: {mat[:, int(cols / 2):]}")
# 4. "Đặt?" Sorry Mr.Hoang, I don't know what you mean so I'm asmming it is "Replace"
# Replace the first element of each row by 0
mat[:, 0] = 0
print(f"4. Dong dau tien cua mat thay bang 0: {mat}")
# 5. Split mat into 2 matrices by column
# ref: https://stackoverflow.com/questions/42190783/how-to-split-a-matrix-into-two-matrices-in-python
mat1, mat2 = np.hsplit(mat, 2)
print(f"5. Tach mat thanh 2 ma tran theo chieu doc: {mat1}, {mat2}")
# 6. Split mat into 2 matrices by row, first one has 1 row, second one has 2 rows
mat1, mat2 = np.vsplit(mat, [1])
print(f"6. Tach mat thanh 2 ma tran theo chieu ngang: {mat1}, {mat2}")
```

## Câu 3 (1 đ)

Tạo `arr` là một tensor 3D kích thước 2x3x4, gồm các số chia đều từ 0 đến 1.

1. Đổi dạng `arr` thành vector.

1. Đổi dạng `arr` thành ma trận có 3 cột.

1. Đổi dạng `arr` thành 3x4x2.

1. Thêm một chiều cho `arr` thành tensor 4D 1x2x3x4.

```python
# Create an tensor 4D with shape 2x3x4 named arr with random values from 0 to 1
arr = np.random.rand(2, 3, 4)
print(f"arr: {arr}")

# 1. Convert arr to a vector
# ref: https://stackoverflow.com/questions/18691084/what-does-1-mean-in-numpy-reshape
arr = arr.reshape(-1)
print(f"1. Chuyen arr thanh vector: {arr}")

# 2. Convert arr to a matrix with 3 columns
arr = arr.reshape(-1, 3)
print(f"2. Chuyen arr thanh ma tran 3 cot: {arr}")

# 3. Convert arr to a 3D tensor with shape 3x4x2
arr = arr.reshape(3, 4, 2)
print(f"3. Chuyen arr thanh 3D tensor: {arr}")

# 4. Add one more dimension to arr to make it 4d with shape 1x2x3x4
arr = arr.reshape(1, 2, 3, 4)
print(f"4. Them 1 chieu vao arr thanh 4D tensor: {arr}")
```

```python
# Get iris dataset
# ref: https://stackoverflow.com/questions/38105539/how-to-convert-a-scikit-learn-dataset-to-a-pandas-dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv(url, names=column_names)

# 1. Subset the iris dataset with petal_width > 1.8 and petal_length < 5.5
# ref: https://stackoverflow.com/questions/11869910/pandas-filter-rows-of-dataframe-with-operator-chaining
iris_subset = iris[(iris['petal_width'] > 1.8) & (iris['petal_length'] < 5.5)]

# 2. Subset the iris dataset with petal_width > average of petal_width
average_petal_width = iris['petal_width'].mean()
iris_subset = iris[iris['petal_width'] > average_petal_width]

# 3. Calculate the min, max, range(max-min), mean each column of iris dataset
sepal_length_min = iris['sepal_length'].min()
sepal_length_max = iris['sepal_length'].max()
sepal_length_range = sepal_length_max - sepal_length_min
sepal_length_mean = iris['sepal_length'].mean()

sepal_width_min = iris['sepal_width'].min()
sepal_width_max = iris['sepal_width'].max()
sepal_width_range = sepal_width_max - sepal_width_min
sepal_width_mean = iris['sepal_width'].mean()

petal_length_min = iris['petal_length'].min()
petal_length_max = iris['petal_length'].max()
petal_length_range = petal_length_max - petal_length_min
petal_length_mean = iris['petal_length'].mean()

petal_width_min = iris['petal_width'].min()
petal_width_max = iris['petal_width'].max()
petal_width_range = petal_width_max - petal_width_min
petal_width_mean = iris['petal_width'].mean()

# 4. Find the row which has petal_witdh_min in iris dataset
iris_subset = iris[iris['petal_width'] == petal_width_min]

# 5. Count the flower species have petal_width > 1.5
iris_subset = iris[iris['petal_width'] > 1.5]
total_flower_species = len(iris_subset['species'].unique())
print(f"5. So loai hoa co petal_width > 1.5: {total_flower_species}")

# 6. Count the flower species have petal_width > 1.5 and species is setosa
iris_subset = iris[(iris['petal_width'] > 1.5) & (iris['species'] == 'Iris-setosa')]
total_flower_species = len(iris_subset['species'].unique())
print(f"6. So luong hoa Iris giong setosa co petal_width lon hon 1.5: {total_flower_species}")

# 7. Calculate min, max, mean of petal_width of flower species is setosa
iris_subset = iris[iris['species'] == 'Iris-setosa']
petal_width_min = iris_subset['petal_width'].min()
petal_width_max = iris_subset['petal_width'].max()
petal_width_mean = iris_subset['petal_width'].mean()
print(f"7. Min, max, mean cua petal_width cua hoa Iris giong setosa: {petal_width_min}, {petal_width_max}, {petal_width_mean}")
```

## Câu 6 (2.5 đ)

**Ảnh** (image) là các ví dụ điển hình của mảng. Ảnh màu RGB là mảng 3D gồm 3 kênh màu Red, Green, Blue. **Ảnh xám** (grayscale) là mảng 2D với các giá trị thường là số nguyên trong khoảng [0, 255] hay số thực trong khoảng [0, 1] (0 là đen nhất, 1 là trắng nhất).

Trong Python, có thể sử dụng thư viện Pillow (https://pypi.org/project/pillow/) để xử lý ảnh và Matplotlib (https://matplotlib.org/) để hiển thị ảnh.

```python
from PIL import Image
import matplotlib.pyplot as plt
```

Bài tập này yêu cầu dùng Numpy để thực hiện vài thao tác **xử lý ảnh** (image processing) bằng cách xử lý mảng `img` được tạo từ ảnh dưới đây.

```python
img = np.asarray(Image.open("image.jpeg").convert("L")) / 255
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()

# My custiom function to show the original image and the result image
def show_result(image, title):    
    plt.figure(figsize=(10, 4))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap="gray")
    plt.axis("off")
    plt.title("Original Image")

    # Result image
    plt.subplot(1, 2, 2)
    plt.imshow(image, cmap="gray")
    plt.axis("off")
    plt.title(title)

    plt.show()
```

```python
print(img.shape)
print(img.min(), img.max(), img.mean())
print(img[:5, :5])
```

### Yêu cầu 1

Tạo ảnh âm bản (negative image)

```python
# Print the negative of img
# ref: https://stackoverflow.com/questions/10965417/how-to-convert-a-numpy-array-to-pil-image-applying-matplotlib-colormap
img_neg = 1 - img

show_result(img_neg, "Negative Image")
```

### Yêu cầu 2

Lật ảnh ngang (horizontal flip)

```python
# Flip img horizontally
# We also can use np.fliplr(img), or np.flip(img, axis=1) to flip horizontally
# And we can use np.flipud(img), or np.flip(img, axis=0) to flip vertically
# This is just another way to flip horizontally, we reverse the order of the columns from left to right
img_flip = img[:, ::-1]

show_result(img_flip, "Flipped Image")
```

### Yêu cầu 3

Xén ảnh (crop)

```python
# Crop the center by 80 percent of the original size
# Explain: We crop 10% of the image from the top, 10% from the bottom, 10% from the left and 10% from the right
# So we crop 20% from the top and bottom, 20% from the left and right
# So we crop 80% from the center
img_crop = img[int(img.shape[0] * 0.1):int(img.shape[0] * 0.9), int(img.shape[1] * 0.1):int(img.shape[1] * 0.9)]

show_result(img_crop, "Cropped Image")
```

### Yêu cầu 4

Thay đổi độ sáng (brightness)

```python
# Increase the brightness of img by 50%
# clip: https://stackoverflow.com/questions/40965066/what-does-numpy-clip-do
brightness_factor = 1.5
img_bright = np.clip(img * brightness_factor, 0, 1)

show_result(img_bright, "Brightened Image")
```

### Yêu cầu 5

Thay đổi độ tương phản (contrast)

```python
# Increase the contrast of img by 70%
contrast_factor = 1.7
img_contrast = np.clip(contrast_factor * (img - 0.5) + 0.5, 0, 1)

show_result(img_contrast, "Contrasted Image")
```
