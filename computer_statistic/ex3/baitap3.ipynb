### Câu 1 (2 điểm)

Bảng dữ liệu sau ghi nhận chế độ dinh dưỡng của 2 nhóm đối tượng (tính theo ounce cho 1 người trong 1 tuần).

|             | Nhóm A | Nhóm B |
|-------------|-------:|-------:|
| White bread | 26	   | 12.5   |
| Sugar	      | 11.5   | 8      |
| Potatoes	  | 48.5   | 33.5   |
| Fruit	      | 13	   | 25.5   |
| Vegetables  | 21.5   | 30.5   |
| Brown bread | 5.0	   | 8      |

1. Vẽ biểu đồ quạt (pie chart) cho nhóm A.
1. Vẽ biểu đồ thanh (bar chart) cho nhóm B.
1. Vẽ biểu đồ so sánh 2 nhóm.
1. Nhận xét.


```python

# Load the data
keys = ['White bread', 'Sugar', 'Potatos', 'Fruit', 'Vegetables', 'Brown bread']
group_a = [26, 11.5, 48.5, 13, 21.5, 5.0]
group_b = [12.5, 8, 33.5, 25.5, 30.5, 8]
df = pd.DataFrame({'keys': keys, 'group_a': group_a, 'group_b': group_b})

# Pie chart for group A
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.pie(df['group_a'], labels=df['keys'], autopct='%1.2f%%', startangle=140)
plt.axis('equal')
plt.title('Figure 1.1: Pie chart for group A')

# Bar chart for group B
plt.subplot(1, 2, 2)
plt.bar(df['keys'], df['group_b'])
plt.title('Figure 1.2: Bar chart for group B')
plt.ylabel('Consumption')
plt.xticks(rotation=45)

# Compare the two groups using a muiltiple bars chart
plt.figure(figsize=(14, 5))
barWidth = 0.35
r1 = np.arange(len(df['keys']))
r2 = [x + barWidth for x in r1]
plt.bar(r1, df['group_a'], width=barWidth, label='Group A')
plt.bar(r2, df['group_b'], width=barWidth, label='Group B')
plt.xlabel('Food')
plt.ylabel('Consumption')
plt.xticks([r + barWidth/2 for r in range(len(df['keys']))], df['keys'])
plt.legend()
plt.title('Figure 1.3: Comparison of the two groups')
plt.xticks(rotation=45)

plt.show()

# Comment
print('Nhận xét: ')
print('Nhóm A tiêu thụ nhiều hơn nhóm B các loại thực phẩm: Bánh mì trắng, Đường, Khoai tây, Rau củ, Bánh mì ngũ cốc')
print('Nhóm B tiêu thụ nhiều hơn nhóm A các loại thực phẩm: Trái cây, Rau củ')
print('Cả hai nhóm tiêu thụ tương đối như nhau các loại thực phẩm: Bánh mì ngũ cốc')
print('Cả hai nhóm đều tiêu tụ nhiều nhất loại thực phẩm: Khoai tây')

```

### Câu 2 (2 điểm)

Bảng dữ liệu sau thống kê số lượng sinh viên có điểm thi trong các khoảng tương ứng của một kì thi (điểm theo thang 100).

| Khoảng điểm |	< 25 | [25, 30) | [30, 40) | [40, 60) | [60, 75) | [75, 100] |
|-------------|-----:|---------:|---------:|---------:|---------:|----------:|
| Số lượng	  | 2	 | 20	    | 56	   | 140	  | 122      | 95        |

1. Lập bảng cho thấy tần số (frequency), tần suất (relative frequency), mật độ (density), tần số tích lũy (cumulative frequency), tần suất tích lũy (cumulative relative frequency).
2. Vẽ biểu đồ histogram.
3. Vẽ biểu đồ tần suất tích lũy.
4. Nhận xét.

```python
# Load the data
# keys= ['< 25', '[25, 30)', '[30, 40)', '[40, 60)', '[60, 75)', '[75, 100]']
keys= ['25', '30', '40', '60', '75', '100']
edges = [0, 25, 30, 40, 60, 75, 100]
frequency= [2, 20, 56, 140, 122, 95]
widths = np.diff(edges)
df = pd.DataFrame({'grade_range': keys, 'frequency': frequency})

# Calculate the relative frequency
df['relative_frequency'] = df['frequency'] / df['frequency'].sum()
# Calculate the density

df['density'] = df['relative_frequency'] / widths
# Calculate the cumulative frequency
df['cumulative_frequency'] = df['frequency'].cumsum()
# Calculate the cumulative relative frequency
df['cumulative_relative_frequency'] = df['relative_frequency'].cumsum()
print(df)

# Plot the histogram
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.bar(df['grade_range'], df['frequency'])
plt.title('Figure 2.1: Histogram of the grades')
plt.ylabel('Frequency')
plt.xlabel('Grade range')
plt.xticks(rotation=45)


# Plot the density
plt.subplot(1, 2, 2)
plt.bar(df['grade_range'], df['density'])
plt.title('Figure 2.2: Density of the grades')
plt.ylabel('Density')
plt.xlabel('Grade range')
plt.xticks(rotation=45)
plt.show()

# Comment
print("Từ biểu đồ histogram, ta có thể thấy rằng đa số sinh viên tập trung ở khoảng điểm từ 40 đến 75.")
```

##### Nhan xet
Phân bố điểm của sinh viên: từ biểu đồ histogram, ta có thể thấy rằng đa số sinh viên tập trung ở khoảng điểm từ 40 đến 75.

### Câu 3 (2 điểm)

Sau đây là số liệu điểm thi của một lớp (tính theo thang 20): 2, 3, 7, 7, 8, 8, 9, 10, 11, 11, 12, 12, 14, 14, 15, 17, 17, 18, 19, 20.

1. Tính trung bình (mean), trung vị (median), yếu vị (mode), phân vị (percentile) 10%, 90%.
2. Tính phạm vi (range), khoảng tứ phân vị (IQR - interquartile range), phương sai (variance), độ lệch chuẩn (standard deviation).
3. Nhận xét.

```python
# Load the data
grades = [2, 3, 7, 7, 8, 8, 9, 10, 11, 11, 12, 12, 14, 14, 15, 17, 17, 18, 19, 20]
df = pd.DataFrame({'grades': grades})

# Print the results
print("Mean:", df.grades.mean())
print("Median:", df.grades.median())
print("Mode:", df.grades.mode().values)
print("10th percentile:", df.grades.quantile(0.1))
print("90th percentile:", df.grades.quantile(0.9))
print("Range:", df.grades.max() - df.grades.min())
lower_quartile, upper_quartile = np.percentile(df.grades, [25, 75])
print("IQR:", upper_quartile - lower_quartile)
print("Variance:", df.grades.var())
print("Standard deviation:", df.grades.std())
```
