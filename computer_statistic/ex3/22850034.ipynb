|             | Nhóm A | Nhóm B |

```python

# Load the data
foods = ['White bread', 'Sugar', 'Potatos', 'Fruit', 'Vegetables', 'Brown bread']
consumption_a = [26, 11.5, 48.5, 13, 21.5, 5.0]
consumption_b = [12.5, 8, 33.5, 25.5, 30.5, 8]
df = pd.DataFrame({'foods': foods, 'consumption_a': consumption_a, 'consumption_b': consumption_b})

# Pie chart for group A
plt.figure(figsize=(10, 5))
plt.pie(df['consumption_a'], labels=df['foods'], autopct='%1.0f%%', startangle=140)
plt.axis('equal')
plt.title('Figure 1.1: Pie chart for group A')

# Bar chart for group B
plt.figure(figsize=(10, 5))
plt.bar(df['foods'], df['consumption_b'], width=barWidth)
plt.title('Figure 1.2: Bar chart for group B')
plt.ylabel('Consumption')
plt.xticks(rotation=45)

# Compare the two groups using a muiltiple bars chart
plt.figure(figsize=(10, 5))
r1 = np.arange(len(df['foods']))
r2 = [x + barWidth for x in r1]
plt.bar(r1, df['consumption_a'], width=barWidth, label='Group A')
plt.bar(r2, df['consumption_b'], width=barWidth, label='Group B')
plt.xlabel('Food')
plt.ylabel('Consumption')
plt.xticks([r + barWidth/2 for r in range(len(df['foods']))], df['foods'])
plt.legend()
plt.title('Figure 1.3: Comparison of the two groups')
plt.xticks(rotation=45)

plt.show()

# Comment
print('Nhận xét:')
print('Nhìn vào biểu đồ 1.1, ta thấy rằng nhóm A tiêu thụ nhiều nhất là khoai tây, tiếp theo là bánh mì trắng và rau cải.')
print('Tiêu thụ ít nhất là bánh mì ngũ cốc và đường.')
print('Cho nên nhóm A là nhóm người chú trọng vào việc ăn uống lành mạnh.')

print('Nhìn vào biểu đồ 1.2, ta thấy rằng nhóm B tiêu thụ nhiều nhất là rau cải, tiếp theo là hoa quả và khoai tây.')
print('Tiêu thụ ít nhất là bánh mì trắng và đường.')
print('Nhóm B là nhóm người chú trọng vào việc ăn uống lành mạnh.')

print('Nhìn vào biểu đồ 1.3, ta thấy rằng nhóm A tiêu thụ nhiều hơn nhóm B ở khoai tây, bánh mì trắng và đường.')
print('Nhóm B tiêu thụ nhiều hơn nhóm A ở hoa quả, rau cải và bánh mì ngũ cốc.')
print('Có thể suy đoán nhóm A chú trọng vào việc ăn uống lành mạnh hơn so với nhóm B.')

```

### Câu 2 (2 điểm)

Bảng dữ liệu sau thống kê số lượng sinh viên có điểm thi trong các khoảng tương ứng của một kì thi (điểm theo thang 100).

| Khoảng điểm |	< 25 | [25, 30) | [30, 40) | [40, 60) | [60, 75) | [75, 100] |
|-------------|-----:|---------:|---------:|---------:|---------:|----------:|
| Số lượng	  | 2	 | 20	    | 56	   | 140	  | 122      | 95        |

1. Lập bảng cho thấy tần số (frequency), tần suất (relative frequency), mật độ (density), tần số tích lũy (cumulative frequency), tần suất tích lũy (cumulative relative frequency).
2. Vẽ biểu đồ histogram.
3. Vẽ biểu đồ tần suất tích lũy.
4. Nhận xét.

```python
# Load the data
data = {'score_range': ['< 25', '[25, 30)', '[30, 40)', '[40, 60)', '[60, 75)', '[75, 100]'],
        'frequency': [2, 20, 56, 140, 122, 95]}

bins = [0, 25, 30, 40, 60, 75, 100]
bin_width = np.diff(bins)
# Create dummy score data
score = [(bins[i] + bins[i + 1]) / 2 for i in range(len(bins) - 1)]

# Create a DataFrame
df = pd.DataFrame(data)

# Relative frequency
total_students = df['frequency'].sum()
df['relative_frequency'] = df['frequency'] / total_students

# Density
df['density'] = df['relative_frequency'] / bin_width

# Cumulative frequency and relative frequency
df['cumulative_frequency'] = df['frequency'].cumsum()
df['cumulative_relative_frequency'] = df['relative_frequency'].cumsum()

# Print the result table
print(df)

# Histogram plot
plt.figure(figsize=(10, 5))
plt.hist(score, bins=bins, weights=df['frequency'], density=True)
plt.title('Histogram of the score data')
plt.xlabel('Score')
plt.ylabel('Frequency')

# Cumulative frequency plot
plt.figure(figsize=(10, 5))
plt.hist(score, bins=bins, weights=df['frequency'], density=True, cumulative=True)
plt.title('Cumulative frequency plot')
plt.xlabel('Score range')
plt.ylabel('Cumulative frequency')

plt.show()

# Comment
print("1. Phân phối điểm:")
print("- Điểm tập trung chủ yếu ở khoảng [40, 60) và [60, 75), với số lượng lần lượt là 140 và 122, chiếm tỉ lệ cao nhất.")
print("- Rất ít sinh viên có điểm dưới 25, chỉ với 2 người, cho thấy tỉ lệ rớt là thấp.")
print("- Có 95 sinh viên đạt điểm từ 75 trở lên, thể hiện tỉ lệ sinh viên có thành tích cao không nhỏ.\n")

print("2. Tần suất tương đối:")
print("- Khoảng [40, 60) có tần suất tương đối cao nhất là 32%, tiếp theo là [60, 75) với 28%.")
print("- Điều này cho thấy hơn một nửa số sinh viên, cụ thể là 60%, có điểm số trong hai khoảng này.\n")

print("3. Mật độ:")
print("- Mật độ cao nhất thuộc về khoảng [60, 75) với 0.018697, theo sau là [40, 60) với 0.016092.")
print("- Mật độ cao phản ánh sự tập trung lớn của sinh viên trong một khoảng điểm nhất định, đặc biệt là điểm cao.\n")

print("4. Tần suất tích lũy và tần suất tích lũy tương đối:")
print("- Chỉ có 2 sinh viên có điểm dưới 25, và tổng cộng 435 sinh viên được tính đến khi đạt tới khoảng điểm cao nhất.")
print("- Tần suất tích lũy tương đối đạt 100% khi tất cả các khoảng điểm đã được tính, cho thấy sự phân phối đầy đủ qua các khoảng.")
```

### Câu 3 (2 điểm)

Sau đây là số liệu điểm thi của một lớp (tính theo thang 20): 2, 3, 7, 7, 8, 8, 9, 10, 11, 11, 12, 12, 14, 14, 15, 17, 17, 18, 19, 20.

1. Tính trung bình (mean), trung vị (median), yếu vị (mode), phân vị (percentile) 10%, 90%.
2. Tính phạm vi (range), khoảng tứ phân vị (IQR - interquartile range), phương sai (variance), độ lệch chuẩn (standard deviation).
3. Nhận xét.

```python
# Load the data
grades = [2, 3, 7, 7, 8, 8, 9, 10, 11, 11, 12, 12, 14, 14, 15, 17, 17, 18, 19, 20]
df = pd.DataFrame({'grades': grades})

# Print the results
print("Mean:", df.grades.mean())
print("Median:", df.grades.median())
print("Mode:", df.grades.mode().values)
print("10th percentile:", df.grades.quantile(0.1))
print("90th percentile:", df.grades.quantile(0.9))
print("Range:", df.grades.max() - df.grades.min())
lower_quartile, upper_quartile = np.percentile(df.grades, [25, 75])
print("IQR:", upper_quartile - lower_quartile)
print("Variance:", df.grades.var())
print("Standard deviation:", df.grades.std())

# Comment
print("1. Trung bình (Mean):")
print(f"- Điểm trung bình của lớp là {df.grades.mean()}, cho thấy mức độ thành tích trung bình của sinh viên nằm ở mức khá, trên trung bình của thang điểm 20.\n")

print("2. Trung vị (Median):")
print(f"- Trung vị là {df.grades.median()}, gần với giá trị trung bình, cho thấy dữ liệu không bị lệch quá nhiều về một phía và phân phối tương đối cân đối.\n")

print("3. Mốt (Mode):")
print(f"- Có nhiều mốt: {df.grades.mode().values}, cho thấy dữ liệu có sự xuất hiện lặp lại của nhiều giá trị khác nhau và không có một số điểm nào nổi bật hơn cả.\n")

print("4. Phân vị thứ 10 và phân vị thứ 90:")
print(f"- 10th percentile là {df.grades.quantile(0.1)}, nghĩa là 10% sinh viên có điểm dưới 6.6.")
print(f"- 90th percentile là {df.grades.quantile(0.9)}, nghĩa là 90% sinh viên có điểm dưới 18.1.")
print("- Khoảng cách giữa hai phân vị này cho thấy phạm vi rộng của điểm số, từ thấp đến cao.\n")

print("5. Phạm vi (Range):")
print(f"- Range là {df.grades.max() - df.grades.min()}, phản ánh sự chênh lệch lớn giữa điểm cao nhất và thấp nhất trong dữ liệu.\n")

print("6. Khoảng tứ phân vị (IQR):")
print(f"- IQR là {upper_quartile - lower_quartile}, cho thấy khoảng cách giữa phân vị thứ 25 và phân vị thứ 75. Điều này cung cấp cái nhìn về sự phân tán của điểm số trung bình, loại bỏ ảnh hưởng của các giá trị ngoại lệ.\n")

print("7. Phương sai (Variance) và Độ lệch chuẩn (Standard Deviation):")
print(f"- Variance là {df.grades.var()}, và Standard Deviation là {df.grades.std()}, cho thấy mức độ phân tán của điểm số xung quanh giá trị trung bình.")
print("- Độ lệch chuẩn khá cao so với thang điểm 20, cho thấy sự chênh lệch lớn về điểm số giữa các sinh viên.\n")
```

### Câu 4 (4 điểm)

```python
# Read data from GDPpercapita.csv
gdp_df = pd.read_csv('GDPpercapita.csv')

# Filter out the GDP data for 2022
gdp_2022 = gdp_df['2022 [YR2022]']

# Filter out countries with missing GDP data in 2022 (2022 [YR2022] column == ..)
gdp_2022 = gdp_2022.loc[gdp_2022 != '..']
# Convert the GDP data to numeric
gdp_2022 = pd.to_numeric(gdp_2022)

gdp_2022_mean = gdp_2022.mean()
gdp_2022_median = gdp_2022.median()
gdp_2022_mode = gdp_2022.mode().values[0]
gdp_2022_min = gdp_2022.min()
gdp_2022_max = gdp_2022.max()
gdp_2022_range = gdp_2022_max - gdp_2022_min
gdp_2022_variance = gdp_2022.var()
gdp_2022_std = gdp_2022.std()
gdp_2022_IQR = gdp_2022.quantile(0.75) - gdp_2022.quantile(0.25)
gdp_2022_10th_percentile = gdp_2022.quantile(0.1)
gdp_2022_90th_percentile = gdp_2022.quantile(0.9)

print("1. PHÂN TÍCH DỮ LIỆU GDP PER CAPITA NĂM 2022")
# Plot top 5 countries with the highest GDP per capita in 2022
sns.barplot(x=gdp_2022.nlargest(5).values, y=gdp_df.loc[gdp_2022.nlargest(5).index, 'Country Name'])
plt.title('Top 5 countries with the highest GDP per capita in 2022')
plt.xlabel('GDP per capita')
plt.ylabel('Country')
plt.show()

# Comment
print("Cả 5 quốc gia có GDP cao nhất đều thuộc khu vực châu Âu, với GDP trên 100,000 USD.")
print("Chứng tỏ khu vực này có nền kinh tế phát triển và thu nhập cao.")

# Plot top 5 countries with the lowest GDP per capita in 2022
sns.barplot(x=gdp_2022.nsmallest(5).values, y=gdp_df.loc[gdp_2022.nsmallest(5).index, 'Country Name'])
plt.title('Top 5 countries with the lowest GDP per capita in 2022')
plt.xlabel('GDP per capita')
plt.ylabel('Country')
plt.show()

# Comment
print("Cả 5 quốc gia có GDP thấp nhất đều thuộc khu vực châu Phi, với GDP dưới 1,000 USD.")
print("Chứng tỏ khu vực này có nền kinh tế yếu đuối và thu nhập thấp.")

# Histogram plot
plt.figure(figsize=(10, 5))
sns.histplot(gdp_2022, bins=30, kde=True)
plt.title('Histogram of GDP per capita in 2022')
plt.xlabel('GDP per capita')
plt.ylabel('Frequency')
plt.show()

# Comment

print("\nNhận xét:")
print(f"- Trung vị GDP là ${gdp_2022_median:,.0f} thấp hơn nhiều so với trung bình ${gdp_2022_mean:,.0f}, cho thấy sự phân bố không đồng đều của GDP. Có một số quốc gia có GDP cao/rất cao làm tăng giá trị trung bình.")
print(f"- Mốt của GDP là ${gdp_2022_mode:,.0f} cho thấy mức GDP thấp nhất cũng là giá trị xuất hiện nhiều nhất trong bộ dữ liệu, có thể do một số quốc gia có kinh tế kém phát triển.")
print(f"- GDP thấp nhất là ${gdp_2022_min:,.0f}, GDP cao nhất là ${gdp_2022_max:,.0f}, phạm vi (Range) GDP là ${gdp_2022_range:,.0f} cho thấy sự chênh lệch rất lớn giữa GDP thấp nhất và cao nhất, phản ánh sự không đồng đều trong phát triển kinh tế toàn cầu.")
print(f"- IQR là ${gdp_2022_IQR:,.0f} cho thấy sự phân tán vừa phải của GDP giữa các quốc gia ở phần giữa bộ dữ liệu, loại trừ các giá trị ngoại lệ ở hai đầu.")
print(f"- Phương sai là ${gdp_2022_variance:,.0f} và độ lệch chuẩn là ${gdp_2022_std:,.0f}: cả hai đều cao, cho thấy sự phân tán lớn của GDP xung quanh giá trị trung bình.")
print("- Độ lệch chuẩn lớn cho thấy sự khác biệt đáng kể về GDP giữa các quốc gia.")
print(f"- 10th percentile là ${gdp_2022_10th_percentile:,.0f} chỉ ra rằng 10% các quốc gia có GDP dưới mức này, thể hiện sự tồn tại của các nền kinh tế nhỏ.")
print(f"- 90th percentile là ${gdp_2022_90th_percentile:,.0f} chỉ ra rằng 90% các quốc gia có GDP dưới mức này, nhấn mạnh sự phân bố không đồng đều và khoảng cách lớn giữa các nền kinh tế lớn và nhỏ.")

# Define the country groups based on the GDP per capita. 
# The thresholds are based on the World Bank classification. (Atlas method, US dollars, 2024)
# I'm using the 2024 thresholds because can't find the 2022 thresholds.
# Reference: https://datahelpdesk.worldbank.org/knowledgebase/articles/906519-world-bank-country-and-lending-groups
bins = [0, 1135, 4465, 13845, float('inf')]
labels = ['Low income', 'Lower middle income', 'Upper middle income', 'High income']
# Create a new column 'income_group' based on the GDP per capita
gdp_2022['income_group'] = pd.cut(gdp_2022, bins=bins, labels=labels)
# Draw pie chart
plt.figure(figsize=(10, 5))
gdp_2022['income_group'].value_counts().plot.pie(autopct='%1.0f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of countries by income group')
plt.xlabel('Income group')
plt.ylabel('Number of countries')
plt.show()

# Comment
print("Phân loại quốc gia theo nhóm thu nhập:")
print("- Nhóm thu nhập cao chiếm tỉ lệ lớn nhất, với 54% số quốc gia.")
print("- Nhóm thu nhập thấp chiếm tỉ lệ thấp nhất, chỉ 12% số quốc gia.")
```

```python
# Filter and clean the data for Vietnam and Thailand
countries = ['Viet Nam', 'Thailand']
gdp_selected = gdp_df[gdp_df['Country Name'].isin(countries)].copy()

# Clear the data with column 'Change', '% Change', and 'income_group' more concisely
columns_to_drop = {'Change', '% Change', 'income_group'}
existing_columns_to_drop = columns_to_drop.intersection(set(gdp_selected.columns))
gdp_selected.drop(columns=existing_columns_to_drop, inplace=True)

# Melting the DataFrame to have a long format
gdp_melted = gdp_selected.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='GDP')
gdp_melted['Year'] = gdp_melted['Year'].str.extract(r'(\d+)')
gdp_melted['GDP'] = pd.to_numeric(gdp_melted['GDP'], errors='coerce')

# Plot
plt.figure(figsize=(10, 5))
for country in countries:
    country_data = gdp_melted[gdp_melted['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GDP'], label=country)

plt.title('GDP per capita of Viet Nam and Thailand from 2003 to 2022')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.legend()
plt.show()

# Comment
print("Biểu đồ cho thấy sự biến động của GDP đầu người của Việt Nam và Thái Lan qua các năm.")
print("Việt Nam có GDP đầu người thấp hơn so với Thái Land nhưng ổn định hơn so với Thái Lan qua các năm.")

# Calculate the GDP growth rate for Vietnam and Thailand
gdp_vietnam = gdp_melted[gdp_melted['Country Name'] == 'Viet Nam'].copy()
gdp_thailand = gdp_melted[gdp_melted['Country Name'] == 'Thailand'].copy()
gdp_vietnam['Change'] = gdp_vietnam['GDP'].diff()
gdp_vietnam['% Change'] = ((gdp_vietnam['Change'] / gdp_vietnam['GDP'].shift(1)) * 100).round(2)
gdp_thailand['Change'] = gdp_thailand['GDP'].diff()
gdp_thailand['% Change'] = ((gdp_thailand['Change'] / gdp_thailand['GDP'].shift(1)) * 100).round(2)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(gdp_vietnam['Year'], gdp_vietnam['% Change'], label='Viet Nam')
plt.plot(gdp_thailand['Year'], gdp_thailand['% Change'], label='Thailand')
plt.title('GDP growth rate of Viet Nam and Thailand from 2003 to 2022')
plt.xlabel('Year')
plt.ylabel('GDP growth rate (%)')
plt.legend()
plt.show()

# Comment
print("Biểu đồ cho thấy sự biến động của tỷ lệ tăng trưởng GDP đầu người của Việt Nam và Thái Lan qua các năm.")
print("Việt Nam có sự biến động lớn hơn so với Thái Lan, với mức tăng trưởng cao nhất vào khoảng năm 2007 và 2010.")
print("Việt Nam tăng trưởng dương qua các năm, trong khi Thái Lan có giảm trưởng trong các năm 2014 và 2020. Trong giai đoạn 2014-2016, Thái Lan có giảm trưởng rất mạnh.")
print("Điều này chỉ ra rằng Việt Nam có tiềm năng phát triển kinh tế mạnh mẽ, và sự chênh lệch GDP per capita giữa hai quốc gia có thể sẽ giảm dần nếu Việt Nam tiếp tục duy trì mức tăng trưởng này.")
```
